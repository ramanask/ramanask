**Linear regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/tvmarketing.csv')

x=np.array(df[['TV']])
y=np.array(df[['Sales']])

plt.scatter(x,y)
plt.xlabel("values of TV")
plt.ylabel("values of Sales")
plt.show()
print(" ")
x_mean=np.mean(x)
y_mean=np.mean(y)
print("Mean values of x_mean is",x_mean,"Mean values of y_mean is",y_mean)
print(" ")

num=0
den=0
for i in range(len(x)):
  num+=(x[i]-x_mean)*(y[i]-y_mean)
  den+=(x[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("intercation is",float(b1),"cofficient is",float(b0))
print(" ")

y_pred=b0+b1*x
y_pred

plt.scatter(x,y,c="red")
plt.plot(x,y_pred)
plt.show()
print(" ")

num_r=0
den_r=0
for i in range(len(x)):
  num_r+=(y[i]-y_pred[i])**2
  den_r+=(x[i]-x_mean)**2
r_sq=1-(num_r/den_r)
print("the values of cofficient determination r_square is",float(r_sq))
```
